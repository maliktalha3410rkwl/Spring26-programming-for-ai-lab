```python
cap1 = int(input("enter jug1 liters: "))
cap2 = int(input("enter jug2 liters: "))
target = int(input("enter target: "))
x = 0
y = 0
stack = [(x, y)]
visited = []
found = False
print("\nSteps:")
while stack:
    x, y = stack.pop()
    if (x, y) in visited:
        continue
    visited.append((x, y))
    if x == target or y == target:
        found = True
        print("found")
        print("Final state: jug1 =", x, "jug2 =", y)
        break
    if x < cap1:
        print("fill jug1")
        stack.append((cap1, y))
    if y < cap2:
        print("fill jug2")
        stack.append((x, cap2))
    if x > 0:
        print("empty jug1")
        stack.append((0, y))
    if y > 0:
        print("empty jug2")
        stack.append((x, 0))
    if x > 0 and y < cap2:
        if x <= cap2 - y:
            pour = x
        else:
            pour = cap2 - y
        print("pour 1 to 2")
        stack.append((x - pour, y + pour))
    if y > 0 and x < cap1:
        if y <= cap1 - x:
            pour = y
        else:
            pour = cap1 - x
        print("pour 2 to 1")
        stack.append((x + pour, y - pour))
if not found:
    print("not found")
```
