```python
n = int(input("Enter number of queens: "))
queens = [-1] * n  
print(f"Enter queen positions (row, column) from 1 to {n}")
for i in range(n):
    r = int(input("row: ")) - 1   
    c = int(input("column: ")) - 1   
    queens[r] = c
valid = True
for i in range(n):
    for j in range(i + 1, n):
        if queens[i] == queens[j]:
            valid = False
        if abs(queens[i] - queens[j]) == abs(i - j):
            valid = False

if valid:
    print("\n correct ")
else:
    print("\n wrong")
horizontal_line = "+" + "----+" * n

for i in range(n):
    print(horizontal_line)
    for j in range(n):
        if queens[i] == j:
            print(f"|{' Q ':^4}", end="")
        else:
            print(f"|{' ':^4}", end="")
    print("|")
print(horizontal_line)
```
